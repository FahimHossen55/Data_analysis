## making all NaN value 0 by fillna()  function 

```python
new_df = df.fillna(0)
```

```python
new_df
```

```python
 new_df = df.fillna({
     'temperature' : 0 ,
     'windspeed' : 0 ,
     'event' : "no event"
 }) 
```

## filling  cell  previous value by fillna

```python
 new_df2 = df.ffill()
```

```python
new_df2
```

## Filling cell with next cell by bfill()

```python
new_df3 = df.bfill()
```

```python
new_df3
```

## we can use axis perameter 

```python
new_df4 = df.fillna( method="bfill", axis="columns") 
new_df4
```

 ## limit

```python
new_df5 = df.fillna( method="ffill", limit=1) 
```
